```python
invoices["Invoice_Date"] = pd.to_datetime(
    invoices["Invoice_Date"],
    format="mixed",
    dayfirst=True,
    errors="coerce"
)

invoices["Due_Date"] = pd.to_datetime(
    invoices["Due_Date"],
    format="mixed",
    dayfirst=True,
    errors="coerce"
)

payments["Payment_Date"] = pd.to_datetime(
    payments["Payment_Date"],
    format="mixed",
    dayfirst=True,
    errors="coerce"
)

sales["Sales_Date"] = pd.to_datetime(
    sales["Sales_Date"],
    format="mixed",
    dayfirst=True,
    errors="coerce"
)
```

# Notebook 1: Data Validation & Cleaning Report
## Project Overview
### Objective

This notebook focuses on validating and preparing the datasets for the Accounts Receivable (AR) & Payment Collection Analysis project.

The objective is to ensure that the data is complete, accurate, and suitable for financial analysis by identifying data quality issues, validating relationships between datasets, and creating analysis-ready files.

The cleaned datasets produced in this notebook will be used for SQL analysis, Python exploratory data analysis (EDA), and Power BI dashboard development.

## Data Quality Assessment

### The following validation checks were performed before analysis:

- Imported all datasets into Python using Pandas.
- Reviewed dataset dimensions and structure.
- Examined data types for all variables.
- Identified missing values.
- Detected duplicate records.
- Removed unnecessary columns.
- Validated relationships between datasets.
- Verified referential integrity.
- Generated summary statistics to identify unusual values.

## Missing Value Treatment

### The missing value assessment identified the following issues:

- The Customer Master dataset contained one unnecessary blank column (Unnamed: 7) generated during the Excel export process. This column was removed as it contained no useful information.
- The Invoice dataset contained six records with missing Invoice_Amount values. Since these values could not be recovered from any related dataset and invoice amount is essential for financial analysis, these records were removed.

After cleaning, all remaining datasets contained no missing values in business-critical fields.

## Duplicate Record Treatment

Duplicate records were evaluated across all datasets.

The validation identified duplicate rows within the Invoice dataset.

These duplicate records were removed to ensure that each invoice represents a unique business transaction.

No duplicate records were found in the remaining datasets.

## Data Type Validation

The data types of each dataset were reviewed and standardized.

### The following conversions were performed:

- Invoice dates were converted to datetime format.
- Due dates were converted to datetime format.
- Payment dates were converted to datetime format.
- Sales dates were converted to datetime format.

Identifier columns such as Customer_ID, Invoice_ID, and Payment_ID were retained as string values to preserve data integrity.

Monetary fields remained numeric to support aggregation and financial calculations.

## Relationship Validation

Referential integrity was validated across all datasets to ensure that related records were accurately linked and suitable for analysis.

The following relationships were verified:

| Relationship | Validation Result |
|--------------|-------------------|
| Customer → Invoice | ✔️ Valid |
| Customer → Sales | ✔️ Valid |
| Customer → Credit | ✔️ Valid |
| Invoice → Payment | ✔️ Valid after data cleaning |

During the validation process, three payment records were found without corresponding invoice records. Further investigation confirmed that these invoices had been removed during the data cleaning stage due to missing `Invoice_Amount` values or duplicate records.

To maintain referential integrity and ensure consistency between the invoice and payment datasets, the three unmatched payment records were excluded from the analysis-ready dataset.

After completing these validation steps, all remaining relationships between the datasets were consistent and reliable for subsequent SQL analysis, exploratory data analysis (EDA), and Power BI reporting.

## Summary Statistics

Summary statistics were generated to verify numerical fields and identify potential anomalies.

Key observations include:

- Payment amounts ranged from 210 to 49,980.
- The average payment amount was approximately 24,072.
- Payment transactions span from January 2024 to December 2026, providing sufficient historical data for trend analysis.
- No abnormal negative payment values or unrealistic financial amounts were identified.

Overall, the financial data appears suitable for analytical modeling and reporting.
